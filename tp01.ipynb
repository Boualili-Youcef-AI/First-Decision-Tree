```python
train, test = da.dataSet_Split(data)
train_accuracies = []
test_accuracies = []

for max_depth in range(30):
    racine = da.Noeud(profondeur_max=max_depth)
    racine.grow(train,0)
    train_accuracies.append(da.precision(racine, train))
    
    test_accuracies.append(da.precision(racine, test))
```

```python
plt.figure(figsize=(10, 6))
depth_values = range(1, len(train_accuracies) + 1)
plt.plot(range(1,31), train_accuracies, label="Pr√©cision Apprentissage", color="blue", marker="o")
plt.plot(range(1,31), test_accuracies, label="Pr√©cision Test", color="red", marker="x")
plt.xlabel("Profondeur maximale")
plt.ylabel("Pr√©cision (%)")
plt.xticks(ticks=[x for x in range(1, len(train_accuracies) + 1, 1)]) 
plt.legend()
plt.grid()
plt.show()
```
